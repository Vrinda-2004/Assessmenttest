```python
import numpy as np
```

```python
array=np.arange(10, 50)
x=array.reshape(4,10)# reshaped into 4 rows and 10 columns
y=x[:,-1]# taking last column from reshaped matrix
y
```

```python
matrix_a = np.full((3, 3), 5) # all elements are 5 , 3 by 3 matrix
matrix_b = np.arange(1, 10).reshape(3, 3) # elements range from 1 to 10, and reshaped into 3 by 3 matrix
matrix_sum = matrix_a + matrix_b # adding the both matrix a and b
matrix_multiply = matrix_a * matrix_b # multiplying both matrices
matrix_transpose = matrix_multiply.T # taking the transpose of the matrix multiplication
matrix_transpose # displaying the transpose matrix
```

```python
arr = np.random.randint(10,101,15)
x = np.where(arr > 50, -1,arr)
count = (arr>50).sum()
count
```

```python
arr = np.random.rand(5, 4)
row_sums = arr.sum(axis=1) ## Find sum of each row
max_row_index = np.argmax(row_sums)
normalized_arr = arr/ arr.sum(axis=1, keepdims=True)
normalized_arr
```

```python
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_elements, counts = np.unique(arr, return_counts=True)
unique_elements
duplicates = unique_elements[counts > 1]
duplicates
```

```python
arr = np.arange(1,37).reshape(6,6)
submatrix = arr[1:5, 1:5]
even = submatrix[submatrix % 2 == 0]
even
```

```python
arr = np.arange(12).reshape(3, 4)
reshaped_array = arr.reshape(2, 6)
flat_array = reshaped_array.flatten()
final_array = flat_array.reshape(4, 3)
final_array
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/titanic_dataset.csv")
data
```

```python
data["Pclass"].value_counts()
```

```python
female_child_survived = data[
    (data["Sex"] == "female") &
    (data["Age"] < 18) &
    (data["Survived"] == 1)
]

female_child_survived
```

```python
median_age = data["Age"].median()
data["Age"].fillna(median_age)
```

```python
def age_category(age):
    if age < 18:
        return "Child"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"

data["age_group"] = data["Age"].apply(age_category)

data[["Age", "age_group"]].head(10)
```

```python
data.sort_values(by="Fare", ascending=False).head(5)
```

```python
data.groupby(["Pclass", "Sex"])["Fare"].mean()
```

```python
data.rename(
    columns={"Sex": "gender","Fare": "ticket_fare"},
    inplace = True
)
data
```
