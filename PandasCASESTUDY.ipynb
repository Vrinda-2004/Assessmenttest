```python
data.rename(columns={'Title':'Movie_Title'}, inplace=True) # inplace is used to update original dataFrame
data.columns
```

```python
data['Genre'].unique() # to get unique genres. unique() is used to remove duplicates and also returns values present in a column
```

```python
data['Genre_Count'] = data['Genre'].str.split(',').str.len()  # instead of str.len() we can also use apply(len)
data[['Movie_Title', 'Genre', 'Genre_Count']]
```

```python
data['Revenue (Millions)'] = data['Revenue (Millions)'].fillna(
    data['Revenue (Millions)'].mean()
    ) # fillna/ Imputation - Replaces missing values with average.
data['Revenue (Millions)']
```

```python
data['Metascore'] = data['Metascore'].fillna(
    data['Metascore'].mean()
)
data['Metascore']
```

```python
data['Profitability'] = (
    data['Revenue (Millions)']
    /
    data['Runtime (Minutes)']
)
data['Profitability']
```

```python
data.nlargest(10,'Profitability')[
    ['Movie_Title','Profitability']
]
```

```python
def category(rating):

    if rating < 6:
        return 'Poor'

    elif rating < 7.5:
        return 'Average'

    else:
        return 'Excellent'

data['Rating_Category'] = data['Rating'].apply(category)
data['Rating_Category'].value_counts()
```

```python
def movie_score(row):

    return (
        row['Rating'] * 10
        + row['Revenue (Millions)']
    )

data['Score'] = data.apply(movie_score, axis=1)
data[['Movie_Title','Score']].head()
```
