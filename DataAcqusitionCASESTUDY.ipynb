DATA TRANSFORMATION

```python
gdp.rename(columns={"Country":"country"}, inplace=True)
gdp
```

```python
literacy.rename(columns={"name":"country"}, inplace=True)
literacy
```

```python
population["population"] = pd.to_numeric(population["population"])
population
```

```python
gdp["GDP"] = pd.to_numeric(gdp["GDP"])
gdp
```

```python
internet["internet_users"] = pd.to_numeric(internet["internet_users"])
internet
```

```python
literacy["literacy_rate"]= pd.to_numeric(literacy["literacy_rate"])
literacy
```

```python
merged = population.merge(gdp,on="country")

merged = merged.merge(internet,on="country")

merged = merged.merge(literacy,on="country")
merged
```

```python
merged["Internet Penetration (%)"] = (
    merged["internet_users"] /
    merged["population"] *100
).round(2)

merged["GDP Per Capita"] = (
    merged["GDP"] /
    merged["population"]
).round(2)

merged.head()
```

```python
conn = sqlite3.connect("country_data.db")

merged.to_sql(
    "country_statistics",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT *
FROM country_statistics
LIMIT 5;
"""

pd.read_sql(query,conn)
```

Data was collected from four different file formats: CSV, Excel, JSON, and XML.
Each dataset was loaded into Pandas for exploration and validation.
Column names were standardized, and numeric fields were converted to appropriate data types.
The datasets were merged using the common country column, and new metrics such as Internet Penetration (%) and GDP Per Capita were calculated.
The cleaned and integrated dataset was stored in a SQLite database named country_statistics.
Analysis showed that the United States had the highest GDP, China had the largest population and the highest number of internet users, the average literacy rate was 97.248%, and there was a strong positive correlation (≈0.773) between literacy rate and internet penetration.
