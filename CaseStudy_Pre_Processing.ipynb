```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/House_Pricing.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.model_selection import train_test_split
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("Duplicate Rows after removal:", df.duplicated().sum())
```

```python
duplicate_columns = []

for i in range(df.shape[1]):
    for j in range(i+1, df.shape[1]):
        if df.iloc[:,i].equals(df.iloc[:,j]):
            duplicate_columns.append(df.columns[j])

print("Duplicate Columns:", duplicate_columns)
```

```python
df = df.drop(columns=duplicate_columns)
df
```

```python
df.isnull().sum()
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
numerical_features = df.select_dtypes(include=np.number).drop(columns=['Sale Price'])
numerical_features.head()
```

```python
minmax = MinMaxScaler()

df_minmax = df.copy()

df_minmax[numerical_features.columns] = minmax.fit_transform(df_minmax[numerical_features.columns])

df_minmax.head()
```

```python
standard = StandardScaler()

df_standard = df.copy()

df_standard[numerical_features.columns] = standard.fit_transform(df_standard[numerical_features.columns])

df_standard.head()
```

```python
categorical_cols = df.select_dtypes(include='object').columns

categorical_cols
```

```python
df_onehot = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

df_onehot.head()
```

```python
label_encoder = LabelEncoder()

df_label = df.copy()

for col in categorical_cols:
    df_label[col] = label_encoder.fit_transform(df_label[col])

df_label.head()
```

```python
Q1 = df[numerical_cols].quantile(0.25)

Q3 = df[numerical_cols].quantile(0.75)

IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
df_no_outliers = df[~((df[numerical_cols] < lower) | (df[numerical_cols] > upper)).any(axis=1)]

print("Original Shape:", df.shape)

print("After Removing Outliers:", df_no_outliers.shape)
```

```python
X = df_onehot.drop("Sale Price", axis=1)

y = df_onehot["Sale Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Training Features:", X_train.shape)

print("Testing Features:", X_test.shape)

print("Training Labels:", y_train.shape)

print("Testing Labels:", y_test.shape)
```
